```python
limitt = 1440
for i in f1.columns:
    for droprate in range(10,limitt,10):
        f1[i+'new'+str(droprate)] = f1[i].shift(droprate) 
        f1[i+'new'+str(droprate)].iloc[0:droprate] = pd.Series(list(f1[i][-droprate:]))
```
