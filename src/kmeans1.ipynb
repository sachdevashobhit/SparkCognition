```python
da1 = pd.read_csv('dryer.csv')
```

```python
da3 = da1.fillna(0)
```

```python
da2 = da3.transpose()
```

```python
da4 = da2.iloc[1:]
```

```python
da5 = da4.transpose()
```

```python
da6 = da5.mean(axis=1)
```

```python
da6.to_csv('dryerproj.csv')
```

```python
from sklearn.cluster import KMeans
k_mea= KMeans(init='k-means++', n_clusters=4)
barc = k_mea.fit(da4)
fdf2 = pd.DataFrame(barc.cluster_centers_)
```

```python
fdf3 = fdf2.transpose()
```
